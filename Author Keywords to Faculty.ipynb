Creating a dictionary from the "faculty_authorkey.txt" text file of faculty(keys) and their author keywords(values).

```python
facukey = open(faculty_authorkey, 'r', encoding="utf8")
read_facukey = facukey.readlines()
facukey_replace = [i.replace('\n', '') for i in read_facukey]
facukey_replace
```

```python
facukey_dict = {}
for each in facukey_replace:
    chunk = each.split(":")
    key = chunk[0]
    value = chunk[1:]
    if key not in set(facukey_dict.keys()):
        facukey_dict[key] = value
    else:
        facukey_dict[key] += value
```

Creating a dictionary from the text file of faculty(keys) and their affiliations(departments).

```python
dept = open(faculty_dept, 'r', encoding="utf8")
read_deptlist = dept.readlines()
dept_replace = [i.replace('\n', '').replace("'", "") for i in read_deptlist]
dept_replace
```

```python
facdept_dict = {}
for each in dept_replace:
    chunk = each.replace(" | ", ":").strip().split(":")
    key = chunk[0]
    value = chunk[1:]
    facdept_dict[key] = value
```

Creating a dictionary linking author keywords(keys) to unique faculty(values) and their de-publicated affiliations(values). 

```python
aukey_dict = {}
missing_keys = []

for a in author_key:
    faculty_list = []
    dept_list = []
    a = a.strip()
    for faculty, aukey in facukey_dict.items():
        if a in aukey:
            faculty_list.append(faculty)
            for fac, dept in facdept_dict.items():
                if faculty in fac:
                    for d in dept:
                        dept_list.append(d)
        else:
            missing_keys.append(a)
            
    joined_faculty = ";".join(list(set(faculty_list)))
    joined_dept = ";".join(list(set(dept_list)))
    
    aukey_dict[a] = [joined_faculty], [joined_dept]
```
